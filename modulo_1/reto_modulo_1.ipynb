```python
import numpy as np
print(np.__version__ )
```

```python
def matriz_de_rotacion(theta:float) -> np.ndarray:
    return np.array([[np.cos(theta), -np.sin(theta)],
                     [np.sin(theta),  np.cos(theta)]])
```

$R(\theta)= \begin{bmatrix} 
\cos\theta & \sin\theta\\ 
-\sin\theta & \cos\theta 
\end{bmatrix}$ 

```python
theta = np.pi / 5
r = matriz_de_rotacion(theta)
print(r)

r_t = np.transpose(r)
print(r_t)

r_i = np.linalg.inv(r)
print(r_i)

r_rt = np.dot(r_t, r)
print(r_rt)

print(np.allclose(np.dot(r, r_t), np.eye(2)))  # De
```
