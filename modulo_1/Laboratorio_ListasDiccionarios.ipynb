# **Introduccin a listas y diccionarios**

##**1. Qu son las listas?**

Las listas son una estructura de datos nativa de python que permite almacenar informacin de manera ordenada y sencilla. Una lista puede almacenar elementos de manera ordenada, donde sus elementos se pueden editar y repetir.

Las listas se pueden definir en python con un par de corchetes cuadrados que almacenan los datos. Una lista con los nmeros de 1 a 5 puede ser creada de la siguiente manera:

```
lista = [1,2,3,4,5]
```
Para obtener la cantidad de elementos dentro de la lista se utiliza la funcin **len** de la siguiente manera:
```
len(lista)
```


**1.1. Indexacin de listas**

Para acceder a un elemento dentro de la lista se debe "indexar" el elemento mediante su posicin. Es importante recordar que la primera posicin de una lista es la posicin 0. As, si se quiere extraer el tercer elemento de la lista definida anteriormente, se debe hacer lo siguiente:
```
lista[2]
```
Note que el tercer elemento corresponde a la posicin 2 dentro de la lista debido a que el primer elemento est en la posicin 0, el segundo elemento en la posicin 1 y el tercer elemento en la posicin 2.

Cuando se quiere acceder a distintos elementos dentro de la lista se debe hacer uso de dos puntos ":" para delimitar los elementos que se quieren extraer. Asuma que quiere extraer los elementos desde el segundo (ndice 1) hasta el cuarto (ndice 3) de la lista definida anteriormente, el cdigo que realiza la operacin mencionada es el siguiente:

```
lista[1:4]
```
Note que en este caso el ltimo ndice es 4 y no 3, esto se debe a que cuando se indexa una lista de esta manera, la indexacin retorna los elementos desde el inicial hasta el final menos uno.

**1.2. Mtodos sobre listas**

Algunos de los mtodos ms utilizados y sus respectivas acciones se definen a continuacin:

**append:** Este mtodo aade el elemento dado por parmetro al final de la lista.

**index:** Este mtodo retorna la posicin del elemento dado por parmetro, en caso de que el elemento dado no se encuentre en la lista, retorna None.

**pop:** Este mtodo remueve el elemento en el ndice dado por parmetro y lo devuelve.

**remove:** Este mtodo remueve el elemento dado por parmetro.

**sort:** Este mtodo devuelve la lista ordenada.

##**2. Qu son los diccionarios?**

Los diccionarios son estructuras de datos nativas de python que permiten almacenar informacin mediante el uso de parejas llave - valor *(key,value)*. Dichos elementos pueden ser modificados pero sus llaves no pueden ser repetidas. A partir de python 3.7 los diccionarios son ordenados.

Los diccionarios en python se definen con corchetes que almacenan las parejas llave valor separadas por coma (,). Un diccionario con la ubicacin del continente de los pases Colombia, Espaa y Canad se define de la siguiente manera:

```
diccionario = {"Colombia":"Sudamrica","Espaa":"Europa","Canad":"Norteamrica"}
```
Para obtener la cantidad de parejas llave - valor dentro de un diccionario se puede utilizar la funcin *len* de la siguiente manera:
```
len(diccionario)
```


**2.1. Indexacin de diccionarios**

Si se desea acceder a un valor dentro del diccionario se debe utilizar su respectiva llave entre corchetes cuadrados. Por ejemplo, para acceder al continente de Espaa se realiza lo siguiente:

```
diccionario["Espaa"]
```

**2.2. Mtodos sobre diccionarios**

Algunos de los mtodos ms utilizados en diccionarios y su funcin se muestran a continuacin:

**keys:** Este mtodo retorna todas las llaves del diccionario en un formato iterable.

**values:** Este mtodo retorna todos los valores del diccionario en un formato iterable.

**items:** Este mtodo retorna todas las parejas llave-valor dentro de tuplas en un formato iterable.

**get:** Este mtodo retorna el valor de la llave ingresada por parmetro, en caso de que no exista la llave dentro del diccionario retorna None.
