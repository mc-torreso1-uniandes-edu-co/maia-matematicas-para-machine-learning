**1.2. Escriba una funcin que reciba como argumento un numero entero positivo y retorne el factorial del mismo.**

**Nota:** Recuerde que el factorial de un nmero est dado por la siguiente frmula:

\begin{align}
        n! = n * (n-1) * (n-2) * (n-3) * ... *(1)
    \end{align}

````python
def factorial(num : int) -> int :
  result = num
  while num > 1 :
    result = result * (num - 1)
    num -= 1

  return result

fact = factorial(4)
print(fact)
````

**1.3. Escriba una funcin que reciba como argumento una lista con nmeros y retorne el promedio los nmeros dentro de la misma.**

````python
def promedio(numeros : list[float]) -> float :
  sum = 0
  for d in numeros:
    sum += d
  return sum / len(numeros)

lista = [25, 32, 65, 32]
p = promedio(lista)
print(p)
````

## **3. Arrays en numpy**

Dentro de NumPy, una de los principales elementos son los *arrays*, una estructura de datos similar a la lista tradicional de Python pero que permite mayor versatilidad a la hora de realizar operaciones. Los arrays son tiles a la hora de representar informacin como vectores y matrices.

Para iniciar un vector con los nmeros de 1 a 5 se utiliza la instruccin 'np.array()' de la siguiente manera:



````python
vector = np.array([1,2,3,4,5])
print(vector)
````

De la misma manera, una matriz puede ser iniciada como una lista de listas segn el nmero de dimensiones que se requiere. Por ejemplo, para iniciar una matriz de 2x2 que contenga los nmeros de 1 a 4 se debe realizar lo siguiente:

````python
matriz = np.array([[1,2],[3,4]])
print(matriz)
print(matriz.shape)
````

Como se ve en la celda anterior, todos los arrays tienen un atributo denominado *shape*, que le indica al usuario las dimensiones del objeto, o lo que es lo mismo, el nmero de objetos en un vector o de filas y columnas en una matriz.

As como los arrays tienen un atributo denominado shape, existen diferentes atributos que dan ms informacin acerca del array. Ac se detallan algunos de los ms comunes:

**array.ndim:** Especifica el nmero de dimensiones de un array. 1 en el caso de un vector, 2 en el caso de una matriz.

**array.size:** Retorna el nmero de elementos dentro de un array.

**array.T:** Retorna el array transpuesto.

Algunos ejemplos de estos atributos se ven a continuacin:

````python
print("Numero de dimensiones en nuestra matriz:", matriz.ndim)
print("Numero de elementos en nuestra matriz:", matriz.size)
print("Matriz transpuesta de nuestra matriz:\n", matriz.T)
````

**3.1. Indexacin de arrays de numpy**


La indexacin es conocida como el proceso para obtener el elemento en una determinada posicin dentro de un arreglo (array, lista). Para ello se utilizan los parntesis cuadrados ([]) para indicar la posicin de la que se quiere extraer el valor. En NumPy la indexacin es similar a la realizada sobre listas con un ligero cambio.

Cuando se tiene un vector cualquier posicin puede ser obtenida simplemente indicando el nombre del vector y la posicin deseada. Por ejemplo:

````python
vector = np.array([1,2,3,4,5])
print(vector[2])
````

Por otra parte, en el caso de las matrices usualmente se realiza con una doble indexacin en el caso de las listas de listas de Python. Esta manera cambia un poco dentro de los arrays, donde se debe indicar dentro de los parntesis cuadrados las "coordenadas" del elemento que queremos. Se debe tener en cuenta que al igual que en las listas de listas la primer coordenada indica la fila mientras que la segunda coordenada indica la columna. Por ejemplo:

````python
matriz = np.array([[1,2],[3,4]])
print(matriz[0,1])
````

De la misma manera, si se desea obtener varios elementos se puede hacer uso de la indexacin con coordenadas en conjunto con el elemento ":" (dos puntos), donde el nmero a la izquierda indica la posicin inicial y el nmero a la derecha indica la posicin final + 1. Por ejemplo:

```
vector[:]  #Devuelve todos los elementos del vector
vector[2:5] #Devuelve los elementos de la segunda a la cuarta posicin
vector[:5] #Devuelve los elementos desde la primera hasta la cuarta posicin.
vector[2:] #Devuelve los elementos a partir de la segunda posicin hasta el final.
vector[::2] #Devuelve los elementos del primero al ltimo con pasos de 2 posiciones.
vector[2:5:2] #Devuelve los elementos de la posicin dos a la posicin cuatro con pasos de 2.
```

En matrices se realiza la misma operacin segn las filas y columnas que se quieren. Por ejemplo:

```
matriz[:,:]  #Devuelve todos los elementos de tanto filas como columnas
matriz[2:5,:] #Devuelve los elementos de la segunda a la cuarta fila y todas las columnas
matriz[:,:5] #Devuelve los elementos para todas las filas desde la primera hasta la cuarta columna.
matriz[2:,:] #Devuelve los elementos a partir de la segunda fila para todas las columnas.
matriz[::2,::3] #Devuelve las filas de la primera a la ltima con pasos de 2 posiciones y de la primera a la ltima columna con pasos de 3 posiciones.
```

1. Escriba una funcin que reciba como argumento una matriz de numpy (array), obtenga todos los elementos de la diagonal de una matriz de NxN y los retorne en un array de NumPy.

````python
m = np.array([[0,1,3], [2,3,4], [4,5,6]])

def diagonal(matrix:np.ndarray) -> np.ndarray :
  return np.diag(matrix)

d = diagonal(m)
print(d)
````

````python
def fila(matrix:np.ndarray, num_row:int) -> np.ndarray :
  return matrix[num_row, :]

f = fila(m, 1)
print(f)

````

3. Escriba una funcin que reciba como argumentos una matriz de NxN y un nmero de columna y devuelva todos los elementos de determinada columna.

**3.3. Tipos especiales de matrices en NumPy**

De igual manera, se pueden crear ciertos tipos de matriz especiales utilizando NumPy mediante el uso del argumento shape visto anteriormente. Algunas matrices pueden ser:

**Matriz donde todos sus elementos son 0:** Esta matriz puede ser creada utilizando la funcin *np.zeros()* donde como argumento *shape* se debe indicar las dimensiones con las que se quiere construir la matriz.

**Matriz donde todos sus elementos son 1:** Esta matriz puede ser creada utilizando la funcin *np.ones()* donde como argumento *shape* se debe indicar las dimensiones con las que se quiere construir la matriz.

**Matriz identidad:** Esta matriz puede ser creada utilizando la funcin *np.eye* donde como argumento obligatorio se debe indicar N, el nmero de filas, y como argumento opcional se puede indicar M (nmero de columnas) y/o k (posicin de la diagonal, donde 0 indica la diagonal principal, un nmero positivo indica una diagonal superior y un nmero negativo indica una diagonal inferior).
